## Identify and Analyze Duplicates


### Task 1: Identify Duplicate Rows
1. Count the number of duplicate rows in the dataset.
3. Display the first few duplicate rows to understand their structure.


```python
# Count the number of duplicate rows
dup_count = df.duplicated().sum()
print(dup_count)
```

```python
# Display the first few duplicate rows
df_dup = df[df.duplicated()]
df_dup.head()
```

### Task 2: Analyze Characteristics of Duplicates
1. Identify duplicate rows based on selected columns such as MainBranch, Employment, and RemoteWork. Analyse which columns frequently contain identical values within these duplicate rows.
2. Analyse the characteristics of rows that are duplicates based on a subset of columns, such as MainBranch, Employment, and RemoteWork. Determine which columns frequently have identical values across these rows.

```python
# Identify duplicate rows based on selected columns
sel_cols = ['MainBranch', 'Employment', 'RemoteWork']
df_dup_sub = df[df.duplicated(subset=sel_cols, keep=False)] # get all duplicate rows

print('Duplicate rows (MainBranch, Employment, RemoteWork):', df_dup_sub.shape[0])
df_dup_sub.head()
```

```python
# Analyse which columns frequently contain identical values within these duplicate rows
for col in sel_cols:
    unique_values = df_dup_sub[col].nunique()
    print(col, ':', unique_values)
```

```python
# Analyse the characteristics of rows that are duplicates based on a subset of columns
grouped_rows = df_dup_sub.groupby(sel_cols).size().reset_index(name='Rows').sort_values(axis=0, by='Rows', ascending=False)
grouped_rows.head()
```

### Task 3: Visualize Duplicates Distribution
1. Create visualizations to show the distribution of duplicates across different categories.
2. Use bar charts or pie charts to represent the distribution of duplicates by Country and Employment.

```python
# Distribution of duplicates across Country
dist_country = df_dup['Country'].value_counts().sort_values(ascending=True)

# Bar chart
dist_country.plot(kind='barh', figsize=(7, 7), color='grey')
plt.title('Distribution of Duplicates by Country')
plt.ylabel('Country')
plt.xlabel('Duplicate Rows')
plt.show()
```

```python
# Distribution of duplicates across Employment
dist_employment = df_dup['Employment'].value_counts().sort_values(ascending=False)

# Pie chart
dist_employment.plot(
    kind='pie',
    labels=None,
    autopct='%1.0f%%',
    startangle=0,
    figsize=(7, 7),
    colors=plt.cm.Set3.colors
)
plt.title('Distribution of Duplicates by Employment')
plt.ylabel('')  # remove default label
plt.legend(labels=dist_employment.index, bbox_to_anchor=(1, 0))
plt.show()
```

### Task 4: Strategic Removal of Duplicates
1. Decide which columns are critical for defining uniqueness in the dataset.
2. Remove duplicates based on a subset of columns if complete row duplication is not a good criterion.

```python
# Decide which columns are critical for defining uniqueness in the dataset
check_cols = pd.DataFrame({
    'Column': df.columns,
    'Unique Values': df.nunique().values,
    'Duplicate Values': [df.duplicated(subset=[col]).sum() for col in df.columns]
})

print('Total Rows:', df.shape[0])
check_cols.sort_values(by='Unique Values', ascending=False, inplace=True)
check_cols[:15]
```

```python
# Remove duplicates based on a subset of columns
df_unique = df.drop_duplicates(subset='ResponseId')
```

```python
# Verify
print(df_unique['ResponseId'].duplicated().sum())
```
